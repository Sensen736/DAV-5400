```python
#import pandas and numpy
import pandas as pd
import numpy as np
```

read file and print

```python
#pd.read_csv() to read file
df = pd.read_csv("https://raw.githubusercontent.com/Sensen736/DAV-5400/master/Tiding_and_Transforming_Data.csv")
# print data
df
```

Drop all the null value and print

```python
#use dropna()to drop null
df = df.dropna(how='all')
df
```

rename the column names with 'Airlines' and 'Status' and print

```python
# use df.rename() to rename the column name
df = df.rename(columns={"Unnamed: 0": "Airlines", "Unnamed: 1": "Status"})
df
```

fill the rest Airline null and print

```python
# use fillna() to fill the null
df['Airlines'].fillna(method='ffill', inplace = True)
df
```

pivot from wide format to long format, use melt() function, and specify the column names and indexs. 

```python
df = df.melt(id_vars = ['Airlines', 'Status'], 
                 value_vars = ['Los Angeles', 'Phoenix', 'San Diego','San Francisco','Seattle'],
                var_name='Citys', value_name='Numbers')
df
```

#3 which airline has the OVERALL on time performance

```python
# store the on time data to ot
ot = df[df['Status'] == 'on time']
#use groupby()to group each airlone and summer the on time number in total
a= ot.groupby('Airlines').sum()
#use groupby()to group each airlone and summer the number in total
a['total']=df.groupby('Airlines').sum()
#calculate the on time rate
a['on_time_rate'] = a['Numbers']/a['total']
#print
a
```
